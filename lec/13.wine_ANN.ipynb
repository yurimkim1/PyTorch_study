훈련데이터와 시험데이터로 분할

```python
train_X, test_X, train_Y, test_Y = train_test_split(wine_data, wine_target, test_size=0.2)
```

훈련데이터 numpy -> tensor로 변환

```python
train_X = torch.from_numpy(train_X).float()
train_Y = torch.from_numpy(train_Y).long()
```

시험데이터 numpy->tensor로 변환

```python
test_X = torch.from_numpy(test_X).float()
test_Y = torch.from_numpy(test_Y).long()
```

설명변수(입력값)와 목적변수(타겟값)의 텐서를 합침

```python
train = TensorDataset(train_X, train_Y)
```

```python
train
```

```python
print(train[0])
```

미니배치로 분할

```python
train_loader = DataLoader(train, batch_size=16, shuffle=True)
```

신경망 구성

```python
class Net(nn.Module):
  def __init__(self):
    super(Net, self).__init__()
    self.fc1 = nn.Linear(13, 96)
    self.fc2 = nn.Linear(96, 2)

  def forward(self, x):
    x = F.relu(self.fc1(x))
    x = self.fc2(x)
    return F.log_softmax(x)
```

```python
model = Net()
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

학습을 시키자

```python
epochs = 300
for epoch in range(epochs):
  total_loss = 0
  # 분할해 둔 데이터를 꺼내옴
  for train_x, train_y in train_loader:
    # 텐서를 감싸고, 계산과정을 기록하는 역할(deprecated)
    train_x, train_y = Variable(train_x), Variable(train_y)
    optimizer.zero_grad()  # 기울기 초기화, 반복할때마다 다시계산
    output = model(train_x)  # model.forward(train_x)와 같음   , 순전파
    loss = criterion(output, train_y) # 오차값 도출
    loss.backward()                   # 오차역전파, 기울기 계산
    optimizer.step()                  # 기울기, lr, 경사하강법으로 w, b를 갱신
    total_loss += loss.data.item()        # 누적 오차 계산
  
  # 50회마다 출력
  if (epoch+1) % 50 == 0:
    print("epoch {} / total_loss {}".format(epoch+1, total_loss))
```
